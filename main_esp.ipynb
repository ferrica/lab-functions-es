## 2. Ahora escribe una funci√≥n que devuelva el elemento m√°s

```python
def greatest(arr):
    if not arr:  # gesti√≥n de lista vac√≠a
        return None
    
    valor_max = arr[0]  # Tomamos el primer elemento como referencia
    for num in arr:
        if num > valor_max:
            valor_max = num  # Actualizamos si encontramos un n√∫mero mayor
    return valor_max
#print(greatest([3, 1, 7, 5, 9, 2])) 
```

## 6. Escribe una funci√≥n que devuelva el factorial de un n√∫mero.

```python
def factorial(n):

    n_factorial = 1

    if n <0:
        n_factorial = -1
        return n_factorial
    elif n == 0 or n == 1:
        n_factorial = 1
        return n_factorial
    else: # n>1
        while n >1 :
            n_factorial = n_factorial * n
            n -= 1
        return n_factorial
        
    
    
```

```python
# F√≥rmula factorial
# n! = n * (n - 1) *...*1

# Este c√≥digo define una funci√≥n llamada "factorial" que toma una entrada "n". La funci√≥n utiliza un bucle for para iterar a trav√©s del rango de n√∫meros
# desde 1 hasta n+1. Para cada n√∫mero en ese rango, multiplica el valor actual de x por el n√∫mero en el rango. Al final del bucle,
# la funci√≥n devuelve el valor final de x, que ser√° el factorial del n√∫mero de entrada "n".

# El factorial de un entero positivo n es el producto de todos los enteros positivos menores o iguales a n.
# Por ejemplo, el factorial de 6 (escrito "6!") es 6 * 5 * 4 * 3 * 2 * 1 = 720.

# As√≠ que esta funci√≥n toma una entrada de cualquier entero positivo y devuelve el factorial de ese n√∫mero.
```

`NOTE: No podemos usar set. ü§î`

```python
def unique(lst_un):
    lista_unica = []  # Lista con valores √∫nicos
    
    for num in lst_un: #extraemos cada uno de los elementos de la lista
       
        if num not in lista_unica:# Solo agregamos el n√∫mero si a√∫n no est√° en la lista √∫nica
            lista_unica.append(num)
    
    return lista_unica
    
#print(unique([1, 2, 3, 2, 4, 1, 5]))
```

## 8. Escribe una funci√≥n que devuelva la moda de una lista, es decir: el elemento que aparece m√°s veces.

`NOTE: No se debe usar count... üßê`

```python
def mode_counter(arr):
    # determinamos la frecuencia de cada elemento. Sumamos + 1 al contador
    frecuencia = {}  # Diccionario para contar las apariciones de cada n√∫mero
    
    for num in arr:
        if num in frecuencia:
            frecuencia[num] += 1  # Incrementamos la cuenta si ya existe
        else:
            frecuencia[num] = 1  # Si es la primera vez, lo inicializamos en 1
    
    # buscamos el n√∫mero con la frecuencia m√°s alta
    #print("Frecuencias:", frecuencia)
    
    max_frecuencia = 0
    moda = None
    
    for num in frecuencia:
        if frecuencia[num] > max_frecuencia:
            max_frecuencia = frecuencia[num]
            moda = num  # Actualizamos la moda
    
    return moda
# print(mode_counter([1, 2, 3, 2, 4,3, 5]))     
```

```python
#import math
def st_dev(list_sd):
    # Contamos la cantidad de elementos sin usar len()
    Num_elementos = 0
    for _ in list_sd:
        Num_elementos += 1
    
    if Num_elementos == 0:
        return None  # Evita divisi√≥n por cero en caso de lista vac√≠a
    
    # Calcular la suma total sin usar sum()
    suma_total = 0
    for num in list_sd:
        suma_total += num
    
    # Calcular la media
    media = suma_total / Num_elementos

    # Calcular la suma de las diferencias al cuadrado
    suma_diferencias_cuadradas = 0
    for num in list_sd:
        diferencia = num - media
        suma_diferencias_cuadradas += diferencia ** 2  # (num - media)¬≤
    
    # Calcular la varianza
    varianza = suma_diferencias_cuadradas / (Num_elementos-1)
    
    # Calcular la ra√≠z cuadrada de la varianza manualmente (m√©todo de aproximaci√≥n)
    x = varianza
   # desv_est = x ** 0.5 # desviaci√≥n est√°ndar como raiz cuadrada de la varianza
    desv_est = math.sqrt(x) # desviaci√≥n est√°ndar como raiz cuadrada de la varianza
     
    return desv_est  
```

## 11. Escribe una funci√≥n que reciba una cadena de palabras separadas por comas y devuelva una cadena de palabras separadas por comas ordenadas alfab√©ticamente.

`NOTA: Puedes usar sorted pero no split y definitivamente no join! ü§™`

```python
def sort_alpha(string):
    palabras = []  # Listacon las palabras
    palabra_actual = ""  
    longitud_cadena = 0  # Simula la longitud sin usar len()
    
    # Contar caracteres sin len()
    for _ in string:
        longitud_cadena += 1

    # Recorremos la cadena manualmente
    for i in range(longitud_cadena):
        char = string[i]
        if char == ",":  # Si encontramos una coma, almacenamos la palabra
            palabras.append(palabra_actual)
            palabra_actual = ""  # Reiniciamos la palabra actual
        else:
            palabra_actual += char  # Acumulamos caracteres en la palabra actual

    palabras.append(palabra_actual)  # Agregamos la √∫ltima palabra despu√©s de la √∫ltima coma

    # Ordenamos la lista de palabras
    palabras_ordenadas = sorted(palabras)

    # Reconstruimos la cadena SIN `join()`
    resultado = ""
    indice_ordenado = 0  # Contador manual para evitar `len()`
    
    for _ in palabras_ordenadas:
        indice_ordenado += 1

    for i in range(indice_ordenado):
        resultado += palabras_ordenadas[i]  # Agregamos la palabra
        if i < indice_ordenado - 1:  # Agregamos coma solo si no es la √∫ltima palabra
            resultado += ","

    return resultado  # Retornamos la cadena final
print(sort_alpha("invaa,Invaso,IronHack,m√≥vil,folio,boli,coche")) 
```

```python
# Para testear tu funci√≥n 
test_alpha(sort_alpha)
```

## 12. Escribe una funci√≥n para verificar si una contrase√±a dada es fuerte (al menos 8 caracteres, al menos una min√∫scula, al menos una may√∫scula, al menos un n√∫mero y al menos un car√°cter especial). Deber√≠a devolver True si es fuerte y False si no lo es.

```python
def check_pass(password):
    tiene_mayuscula = False
    tiene_minuscula = False
    tiene_numero = False
    tiene_especial = False

    caracteres_especiales = "#@!$%&()^*[]{}"
    
    # contamos n√∫mero de caracateres
    contador = 0
    for _ in password:
        contador += 1

    if contador < 8:  # menos de 8 caracteres no es v√°lido
        print("menos de 8")
        return False

    # Recorrer manualmente la contrase√±a
    for caracter in password:
        if "A" <= caracter <= "Z":  # May√∫scula?
            print("tiene mayusculas")
            tiene_mayuscula = True
        
        elif "a" <= caracter <= "z":  # Min√∫scula?
            print("tiene minuscilas")
            tiene_minuscula = True
        
        elif "0" <= caracter <= "9":  # N√∫mero?
            print("contine n√∫meros")
            tiene_numero = True
        else:
            # Verificar manualmente si es un car√°cter especial
            indice_especial = 0
            for especial in caracteres_especiales:
                if caracter == especial:
                    tiene_especial = True
                    print("tiene_especial")
                indice_especial += 1

    if (tiene_mayuscula and tiene_minuscula and tiene_numero and tiene_especial):
        return True
    else:
        return False
        
print(check_pass("Abc123!@"))
print(check_pass("abcdefg1")) # no tiene may√∫sculas, no dber√≠a darlo por bueno
```

```python
# Para testear tu funci√≥n
test_pass(check_pass)
```
